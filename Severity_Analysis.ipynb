```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from sklearn.utils.class_weight import compute_class_weight
import pandas as pd
import numpy as np
from PIL import Image, ImageFile
from tqdm import tqdm
import os
import matplotlib.pyplot as plt

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Allow loading of truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Load and update image paths from TSV files
def load_and_update_paths(tsv_file, root_folder):
    df = pd.read_csv(tsv_file, sep='\t')
    
    def adjust_path(x):
        full_path = os.path.join(root_folder, x.replace('data/', '').replace('/', '\\'))
        return full_path
    
    df['image_path'] = df['image_path'].apply(adjust_path)
    df = df[df['image_path'].apply(os.path.exists)]
    
    return df

train_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_train_final.tsv"
val_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_dev_final.tsv"
test_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_test_final.tsv"
image_root_folder = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\data"

train_df = load_and_update_paths(train_tsv_file, image_root_folder)
val_df = load_and_update_paths(val_tsv_file, image_root_folder)
test_df = load_and_update_paths(test_tsv_file, image_root_folder)

# Label Mapping
label_mapping = {label: idx for idx, label in enumerate(train_df['class_label'].unique())}
train_df['class_label'] = train_df['class_label'].map(label_mapping)
val_df['class_label'] = val_df['class_label'].map(label_mapping)
test_df['class_label'] = test_df['class_label'].map(label_mapping)

# Dataset class
class DisasterDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['image_path']
        label = self.dataframe.iloc[idx]['class_label']
        
        try:
            image = Image.open(img_path).convert("RGB")
        except OSError:
            print(f"Skipping corrupted image: {img_path}")
            return None

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)

# Class weights
class_weights = compute_class_weight('balanced', classes=np.unique(train_df['class_label']), y=train_df['class_label'])
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

# Data transformations
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Reduced size for a simple CNN
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Data loaders
train_dataset = DisasterDataset(train_df, transform=transform)
val_dataset = DisasterDataset(val_df, transform=transform)
test_dataset = DisasterDataset(test_df, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Define a simple CNN model for image classification
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=3):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(32, 128, kernel_size=3, stride=1, padding=1)        
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.fc1 = nn.Linear(64 * 16 * 16, 128)
        self.fc2 = nn.Linear(128, num_classes)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 64 * 16 * 16)
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Instantiate and transfer the model to the GPU
model = SimpleCNN(num_classes=3).to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.Adam(model.parameters(), lr=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# Lists to store training and validation metrics
train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

# Training function
# Training function with early stopping
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=50):
    best_val_loss = float('inf')  # Initialize best validation loss
    patience_counter = 0
    early_stopping_patience = 5  # Adjust as needed
    best_model_state = None
    
    for epoch in range(num_epochs):
        model.train()
        train_loss, correct, total = 0, 0, 0
        
        for images, labels in tqdm(train_loader, desc=f"Training Epoch {epoch+1}/{num_epochs}"):
            if images is None:
                continue

            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
        
        train_accuracy = 100 * correct / total
        val_loss, val_accuracy = validate_model(model, val_loader, criterion)
        
        # Append metrics for plotting
        train_losses.append(train_loss / len(train_loader))
        val_losses.append(val_loss)
        train_accuracies.append(train_accuracy)
        val_accuracies.append(val_accuracy)
        
        print(f"Epoch [{epoch+1}/{num_epochs}], "
              f"Train Loss: {train_losses[-1]:.4f}, "
              f"Train Accuracy: {train_accuracies[-1]:.2f}%, "
              f"Val Loss: {val_losses[-1]:.4f}, "
              f"Val Accuracy: {val_accuracies[-1]:.2f}%")
        
        scheduler.step(val_loss)
        
        # Early Stopping logic
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0  # Reset counter if validation loss improves
            best_model_state = model.state_dict()  # Save best model state
        else:
            patience_counter += 1
        
        if patience_counter >= early_stopping_patience:
            print(f"Early stopping at epoch {epoch+1}")
            model.load_state_dict(best_model_state)  # Load the best model
            break


# Validation function
def validate_model(model, val_loader, criterion):
    model.eval()
    val_loss, correct, total = 0, 0, 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            if images is None:
                continue
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            val_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
    
    val_accuracy = 100 * correct / total
    return val_loss / len(val_loader), val_accuracy

# Run the training
train_model(model, train_loader, val_loader, criterion, optimizer, scheduler)

# Plotting function for loss and accuracy
def plot_metrics(train_losses, val_losses, train_accuracies, val_accuracies):
    epochs = range(1, len(train_losses) + 1)
    
    # Plot loss
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_losses, 'b', label='Training Loss')
    plt.plot(epochs, val_losses, 'r', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_accuracies, 'b', label='Training Accuracy')
    plt.plot(epochs, val_accuracies, 'r', label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.show()

# Plot the metrics after training
plot_metrics(train_losses, val_losses, train_accuracies, val_accuracies)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from torchvision.models import resnet50, ResNet50_Weights
from sklearn.utils.class_weight import compute_class_weight
import pandas as pd
import numpy as np
from PIL import Image, ImageFile
from tqdm import tqdm
import os
import matplotlib.pyplot as plt
import optuna

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Allow loading of truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Load and update image paths from TSV files
def load_and_update_paths(tsv_file, root_folder):
    df = pd.read_csv(tsv_file, sep='\t')
    def adjust_path(x):
        full_path = os.path.join(root_folder, x.replace('data/', '').replace('/', '\\'))
        return full_path
    df['image_path'] = df['image_path'].apply(adjust_path)
    df = df[df['image_path'].apply(os.path.exists)]
    return df

# Paths to dataset files
train_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_train_final.tsv"
val_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_dev_final.tsv"
test_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_test_final.tsv"
image_root_folder = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\data"

train_df = load_and_update_paths(train_tsv_file, image_root_folder)
val_df = load_and_update_paths(val_tsv_file, image_root_folder)
test_df = load_and_update_paths(test_tsv_file, image_root_folder)

# Label Mapping
label_mapping = {label: idx for idx, label in enumerate(train_df['class_label'].unique())}
train_df['class_label'] = train_df['class_label'].map(label_mapping)
val_df['class_label'] = val_df['class_label'].map(label_mapping)
test_df['class_label'] = test_df['class_label'].map(label_mapping)

# Dataset class
class DisasterDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['image_path']
        label = self.dataframe.iloc[idx]['class_label']
        
        try:
            image = Image.open(img_path).convert("RGB")
        except OSError:
            print(f"Skipping corrupted image: {img_path}")
            return None

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)

# Class weights
class_weights = compute_class_weight('balanced', classes=np.unique(train_df['class_label']), y=train_df['class_label'])
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

# Data transformations
transform = transforms.Compose([
    transforms.RandomResizedCrop((64, 64), scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Data loaders
train_dataset = DisasterDataset(train_df, transform=transform)
val_dataset = DisasterDataset(val_df, transform=transform)
test_dataset = DisasterDataset(test_df, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Define ResNet-50-based model
class ResNet50Model(nn.Module):
    def __init__(self, num_classes=3, dropout_rate=0.5, freeze_base=False):
        super(ResNet50Model, self).__init__()
        weights = ResNet50_Weights.DEFAULT  # Use default weights
        self.resnet = resnet50(weights=weights)
        
        # Freeze layers if specified
        if freeze_base:
            for param in self.resnet.parameters():
                param.requires_grad = False

        # Modify the final fully connected layer with dropout
        in_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(in_features, num_classes)
        )

    def forward(self, x):
        return self.resnet(x)

# Instantiate the ResNet-50 model and move to device
model = ResNet50Model(num_classes=3, dropout_rate=0.5).to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# Training function with Early Stopping
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=100, patience=5):
    best_val_loss = float('inf')
    patience_counter = 0
    train_losses, val_losses, train_accuracies, val_accuracies = [], [], [], []
    
    for epoch in range(num_epochs):
        model.train()
        train_loss, correct, total = 0, 0, 0
        
        for images, labels in tqdm(train_loader, desc=f"Training Epoch {epoch+1}/{num_epochs}"):
            if images is None:
                continue

            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
        
        train_accuracy = 100 * correct / total
        val_loss, val_accuracy = validate_model(model, val_loader, criterion)
        
        train_losses.append(train_loss / len(train_loader))
        val_losses.append(val_loss)
        train_accuracies.append(train_accuracy)
        val_accuracies.append(val_accuracy)
        
        print(f"Epoch [{epoch+1}/{num_epochs}], "
              f"Train Loss: {train_losses[-1]:.4f}, "
              f"Train Accuracy: {train_accuracies[-1]:.2f}%, "
              f"Val Loss: {val_losses[-1]:.4f}, "
              f"Val Accuracy: {val_accuracies[-1]:.2f}%")
        
        scheduler.step(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping triggered.")
                break

# Validation function
def validate_model(model, val_loader, criterion):
    model.eval()
    val_loss, correct, total = 0, 0, 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            if images is None:
                continue
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            val_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
    
    val_accuracy = 100 * correct / total
    return val_loss / len(val_loader), val_accuracy

# Plotting function
def plot_metrics(train_losses, val_losses, train_accuracies, val_accuracies):
    epochs = range(1, len(train_losses) + 1)
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_losses, 'b', label='Training Loss')
    plt.plot(epochs, val_losses, 'r', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_accuracies, 'b', label='Training Accuracy')
    plt.plot(epochs, val_accuracies, 'r', label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.show()

# Hyperparameter optimization with Optuna
def objective(trial):
    lr = trial.suggest_loguniform('lr', 1e-5, 1e-3)
    dropout_rate = trial.suggest_uniform('dropout_rate', 0.3, 0.6)
    
    model = ResNet50Model(num_classes=3, dropout_rate=dropout_rate).to(device)
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5)
    
    criterion = nn.CrossEntropyLoss(weight=class_weights)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)
    
    train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=100, patience=3)
    val_loss, val_accuracy = validate_model(model, val_loader, criterion)
    
    return val_loss

study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=10)

# Save the model
torch.save(model.state_dict(), 'pytorch_severity_classification_model_resnet50.pth')
print("Model saved to pytorch_severity_classification_model_resnet50.pth")
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from sklearn.utils.class_weight import compute_class_weight
import pandas as pd
from PIL import Image, ImageFile
import matplotlib.pyplot as plt
from tqdm import tqdm  # Import tqdm for progress bars

# Allow truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Paths
train_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_train_final.tsv"
val_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_dev_final.tsv"
test_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_test_final.tsv"
image_root_folder = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\data"

# Load and update image paths
def load_and_update_paths(tsv_file, root_folder):
    df = pd.read_csv(tsv_file, sep='\t')
    df['image_path'] = df['image_path'].apply(lambda x: os.path.join(root_folder, x.replace('data/', '').replace('/', '\\')))
    df = df[df['image_path'].apply(os.path.exists)]  # Filter out missing files
    return df

# Load datasets
train_df = load_and_update_paths(train_tsv_file, image_root_folder)
val_df = load_and_update_paths(val_tsv_file, image_root_folder)
test_df = load_and_update_paths(test_tsv_file, image_root_folder)

# Label mapping
label_mapping = {label: idx for idx, label in enumerate(train_df['class_label'].unique())}
train_df['class_label'] = train_df['class_label'].map(label_mapping)
val_df['class_label'] = val_df['class_label'].map(label_mapping)
test_df['class_label'] = test_df['class_label'].map(label_mapping)

# Custom Dataset
class DisasterDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['image_path']
        label = self.dataframe.iloc[idx]['class_label']
        
        try:
            image = Image.open(img_path).convert("RGB")
        except (OSError, FileNotFoundError):
            print(f"Skipping corrupted or missing image: {img_path}")
            return self.__getitem__((idx + 1) % len(self))

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)

# Data transformations
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Datasets and DataLoaders
train_dataset = DisasterDataset(train_df, transform=transform)
val_dataset = DisasterDataset(val_df, transform=transform)
test_dataset = DisasterDataset(test_df, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Class weights for imbalanced datasets
class_weights = compute_class_weight('balanced', classes=list(label_mapping.values()), y=train_df['class_label'])
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

# Model Definition with Fuzzy Logic
class ResNet18WithFuzzyLogic(nn.Module):
    def __init__(self, num_classes):
        super(ResNet18WithFuzzyLogic, self).__init__()
        self.resnet18 = models.resnet18(weights="IMAGENET1K_V1")
        self.resnet18.fc = nn.Linear(self.resnet18.fc.in_features, num_classes)

        # Fuzzy Logic Layer
        self.fuzzy_means = nn.Parameter(torch.randn(num_classes, num_classes))
        self.fuzzy_sigmas = nn.Parameter(torch.ones(num_classes, num_classes) * 0.5)

    def fuzzy_layer(self, x):
        # Gaussian-like fuzzy membership function
        x_expanded = x.unsqueeze(1)  # Shape: [batch_size, 1, num_classes]
        membership_scores = torch.exp(-0.5 * ((x_expanded - self.fuzzy_means) / self.fuzzy_sigmas) ** 2)
        return membership_scores

    def forward(self, x):
        x = self.resnet18(x)  # Standard ResNet18
        fuzzy_scores = self.fuzzy_layer(x)  # Fuzzy logic applied
        return x  # Use logits for classification tasks

# Training and Validation Functions with Early Stopping
def train_and_validate_with_early_stopping(model, train_loader, val_loader, criterion, optimizer, num_epochs, patience):
    train_accuracies = []
    val_accuracies = []
    best_val_loss = float('inf')
    epochs_without_improvement = 0
    best_model_wts = model.state_dict()  # Save best model weights

    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch + 1}/{num_epochs}")

        # Training phase
        model.train()
        train_loss, train_correct, train_total = 0, 0, 0
        train_progress = tqdm(train_loader, desc="Training", ncols=100)
        for batch_idx, (images, labels) in enumerate(train_progress):
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            train_total += labels.size(0)
            train_correct += (predicted == labels).sum().item()

            train_progress.set_postfix(loss=train_loss / (batch_idx + 1), accuracy=100 * train_correct / train_total)

        train_acc = 100 * train_correct / train_total
        train_accuracies.append(train_acc)
        print(f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_acc:.2f}%")

        # Validation phase
        model.eval()
        val_loss, val_correct, val_total = 0, 0, 0
        val_progress = tqdm(val_loader, desc="Validation", ncols=100)
        with torch.no_grad():
            for images, labels in val_progress:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

                val_progress.set_postfix(loss=val_loss / (batch_idx + 1), accuracy=100 * val_correct / val_total)

        val_acc = 100 * val_correct / val_total
        val_accuracies.append(val_acc)
        print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_acc:.2f}%")

        # Early stopping check
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            epochs_without_improvement = 0
            best_model_wts = model.state_dict()  # Save best model weights
        else:
            epochs_without_improvement += 1
            if epochs_without_improvement >= patience:
                print(f"Early stopping triggered! No improvement for {patience} epochs.")
                break

    model.load_state_dict(best_model_wts)  # Restore best model
    return train_accuracies, val_accuracies

# Test function for final accuracy
def test(model, test_loader):
    model.eval()
    test_correct, test_total = 0, 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            test_total += labels.size(0)
            test_correct += (predicted == labels).sum().item()

    test_acc = 100 * test_correct / test_total
    print(f"Test Accuracy: {test_acc:.2f}%")
    return test_acc

# Instantiate model and optimizer
model = ResNet18WithFuzzyLogic(num_classes=len(label_mapping)).to(device)
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train and validate with early stopping
train_accuracies, val_accuracies = train_and_validate_with_early_stopping(
    model, train_loader, val_loader, criterion, optimizer, num_epochs=75, patience=3
)

# Plot the training and validation accuracy
plt.plot(range(1, len(train_accuracies) + 1), train_accuracies, label="Training Accuracy")
plt.plot(range(1, len(val_accuracies) + 1), val_accuracies, label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# Test the model at the end
test(model, test_loader)
```

```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models
from sklearn.utils.class_weight import compute_class_weight
import pandas as pd
from PIL import Image, ImageFile
import albumentations as A
from albumentations.pytorch import ToTensorV2
from tqdm import tqdm
import matplotlib.pyplot as plt

# Allow truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Paths
train_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_train_final.tsv"
val_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_dev_final.tsv"
test_tsv_file = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\consolidated_damage_test_final.tsv"
image_root_folder = r"C:\Users\VIT-AP\Downloads\data_damage_severity\data_damage_severity\data"

# Load and update image paths
def load_and_update_paths(tsv_file, root_folder):
    df = pd.read_csv(tsv_file, sep='\t')
    df['image_path'] = df['image_path'].apply(
        lambda x: os.path.join(root_folder, x.replace('data/', '').replace('/', '\\')))
    df = df[df['image_path'].apply(os.path.exists)]
    return df

train_df = load_and_update_paths(train_tsv_file, image_root_folder)
val_df = load_and_update_paths(val_tsv_file, image_root_folder)
test_df = load_and_update_paths(test_tsv_file, image_root_folder)

# Label mapping
label_mapping = {label: idx for idx, label in enumerate(train_df['class_label'].unique())}
train_df['class_label'] = train_df['class_label'].map(label_mapping)
val_df['class_label'] = val_df['class_label'].map(label_mapping)
test_df['class_label'] = test_df['class_label'].map(label_mapping)

# Custom Dataset
class DisasterDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['image_path']
        label = self.dataframe.iloc[idx]['class_label']
        try:
            image = Image.open(img_path).convert("RGB")
        except (OSError, FileNotFoundError):
            print(f"Skipping corrupted or missing image: {img_path}")
            return self.__getitem__((idx + 1) % len(self))

        if self.transform:
            image = self.transform(image=np.array(image))['image']

        return image, torch.tensor(label, dtype=torch.long)

# Data transformations
train_transform = A.Compose([
    A.RandomResizedCrop(128, 128),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
    A.CoarseDropout(max_holes=8, max_height=16, max_width=16, p=0.5),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])

val_transform = A.Compose([
    A.Resize(128, 128),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])

# Model definition
class ResNetWithDropout(nn.Module):
    def __init__(self, base_model, num_classes, dropout_rate=0.5):
        super(ResNetWithDropout, self).__init__()
        self.base = nn.Sequential(*list(base_model.children())[:-1])
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(base_model.fc.in_features, num_classes)

    def forward(self, x):
        x = self.base(x)
        x = torch.flatten(x, 1)
        x = self.dropout(x)
        x = self.fc(x)
        return x

model = ResNetWithDropout(models.resnet18(pretrained=True), len(train_df['class_label'].unique()))
model = model.to(device)

# Datasets and DataLoaders
train_dataset = DisasterDataset(train_df, transform=train_transform)
val_dataset = DisasterDataset(val_df, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Class weights for imbalanced datasets
class_weights = compute_class_weight('balanced', classes=list(label_mapping.values()), y=train_df['class_label'])
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)

# Training loop
def train_model(model, criterion, optimizer, train_loader, val_loader, scheduler, num_epochs=50, patience=10):
    train_loss, val_loss, train_acc, val_acc = [], [], [], []
    best_val_loss = float('inf')
    epochs_no_improve = 0

    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch + 1}/{num_epochs}")
        print("-" * 50)

        # Training phase
        model.train()
        running_loss, correct_preds, total_preds = 0.0, 0, 0
        for images, labels in tqdm(train_loader, desc="Training", ncols=100):
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            correct_preds += (preds == labels).sum().item()
            total_preds += labels.size(0)

        epoch_loss = running_loss / len(train_loader)
        epoch_acc = (correct_preds / total_preds) * 100
        train_loss.append(epoch_loss)
        train_acc.append(epoch_acc)
        print(f"Train Loss: {epoch_loss:.4f}, Train Accuracy: {epoch_acc:.2f}%")

        # Validation phase
        model.eval()
        running_val_loss, correct_val_preds, val_total_preds = 0.0, 0, 0
        with torch.no_grad():
            for images, labels in tqdm(val_loader, desc="Validation", ncols=100):
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)

                running_val_loss += loss.item()
                correct_val_preds += (torch.argmax(outputs, dim=1) == labels).sum().item()
                val_total_preds += labels.size(0)

        epoch_val_loss = running_val_loss / len(val_loader)
        epoch_val_acc = (correct_val_preds / val_total_preds) * 100
        val_loss.append(epoch_val_loss)
        val_acc.append(epoch_val_acc)
        print(f"Validation Loss: {epoch_val_loss:.4f}, Validation Accuracy: {epoch_val_acc:.2f}%")

        scheduler.step(epoch_val_loss)

        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            epochs_no_improve = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print(f"Early stopping after {patience} epochs of no improvement.")
                break

    # Load the best model
    model.load_state_dict(torch.load('best_model.pth'))
    return train_loss, val_loss, train_acc, val_acc

# Train the model
train_loss, val_loss, train_acc, val_acc = train_model(
    model, criterion, optimizer, train_loader, val_loader, scheduler, num_epochs=50, patience=10
)

# Plot results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Training Loss')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Curve')

plt.subplot(1, 2, 2)
plt.plot(range(1, len(train_acc) + 1), train_acc, label='Training Accuracy')
plt.plot(range(1, len(val_acc) + 1), val_acc, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy Curve')
plt.show()
```
